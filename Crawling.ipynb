```python
from openpyxl import load_workbook
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
# 엑셀 열기
wb = load_workbook(filename = './DoungJi_정보.xlsx')
```

```python
# G열에 있는 모든 텍스트 정보 List로 만들기
ws = wb.active
first_column = ws['G']
address_list = []
for x in range(len(first_column)): 
    address_list.append(first_column[x].value) 
```

```python
# Selenium으로 해당 주소로 접속하기
driver = webdriver.Chrome('./chromedriver.exe')
url = 'http://f.goodkiss.co.kr/naver/navermap.html'
driver.get(url)
```

```python
for idx, addr in enumerate(address_list[3:]):
    try:
        address = addr.replace('\xa0', '')
        driver.find_element_by_name('addr').send_keys(address)
        driver.find_element_by_name('btnSearch').click()
        print(addr)
        source = driver.page_source
        soup = BeautifulSoup(source, 'html.parser')
        try:
            td_list = soup.select('td')
            ws['H'+str(idx+3)] = td_list[4].text
            ws['I'+str(idx+3)] = td_list[5].text
        except:
            ws['H'+str(idx+3)] = ''
            ws['I'+str(idx+3)] = ''
        driver.find_element_by_name('btnSearch').click()
    except:
        pass
wb.save(filename = './DoungJi_정보.xlsx')
```
